```python
import numpy as np

import matplotlib as ml
import matplotlib.pyplot as plt

import ioeddy as io 
import matplotlib.gridspec as gridspec

import pandas as pd

path='/home/jose/WORK/scratch/EDDY_OUTLET_BC/run/'

filename = path+'restartfiles/u_00031000.res'
var = 'w'


i,j,k,it,time,dt,grav,data = io.readres(filename)
```

```python
print('i,j,k=', i,j,k)
print('dtm=',dt)
print('grav=', grav)
print('size=',data.shape)
print('data=',data[1,2,5])
print('max_data=', np.nanmax(data))
print('min_data=', np.nanmin(data))
print('nstep=',it)
```

```python
nx, index, x, xe, xc = io.readgrid(path+'x3_grid.in')
nr, index, r, re, rc = io.readgrid(path+'x1_grid.in')
nth, index, th, the, thc = io.readgrid('./x2_grid_half.in')
```

```python
xedges = thc
xedges[0]=0
yedges = rc[1:-1]
X,Y =np.meshgrid(xedges, yedges)
```

```python
#fixging the azymutal gap by interpolation
data_ng = data[1:-1,:,1:-1]
data_ng[:,0,:] = 0.5*(data_ng[:,1,:]+data_ng[:,-2,:])
data_ng[:,-1,:] = data_ng[:,0,:] 
data_slice = data_ng[:,:,io.loc(xc,0)]
#data_slice = data_ng[:,:,-1]
```

```python
io.loc(xc,20)
```

```python
df_X = pd.DataFrame(X,    index=yedges, columns=xedges)
df_Y = pd.DataFrame(Y,    index=yedges, columns=xedges)
df_U = pd.DataFrame(data_slice, index=yedges, columns=xedges)

    #df_U.shape

    #df_U.head(3)

    #df_U.tail(3)
xmin = 0
xmax = 6.283185
ymin = 0
ymax = 20

roi = (xmin <= df_X) & (df_X <= xmax) & (ymin <= df_Y) & (df_Y <= ymax)

df_X_crop = df_X[roi].dropna(axis=[0,1], how='all')
df_Y_crop = df_Y[roi].dropna(axis=[0,1], how='all')
df_U_crop = df_U[roi].dropna(axis=[0,1], how='all')
```

```python
fig = plt.figure(figsize=(8,2), dpi=300)
gsp = gridspec.GridSpec(1,1)
plt.subplot(gsp[0,0],projection='polar')

#minval=np.min(data_slice)
#maxval=np.max(data_slice)

minval=-0.01
maxval=0.01

plt.pcolormesh(df_X_crop, df_Y_crop, df_U_crop,vmin=minval,vmax=maxval,cmap='seismic')
plt.colorbar()
plt.show()
```
