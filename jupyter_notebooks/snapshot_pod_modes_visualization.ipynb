```python
###.......................Reading the low dimensional mode.....................................

N = 2501       # Number of snapshots

nstart_mode = 2000
nend_mode   = 2501

path = '/home/sheel/Work/projects/spod_re5e4/post/frinf/snapshot_pod_data/x_D_100/frinf_2111900_2501/pod_modes/'
#path = '/home/sheel/Work/projects/spod_re5e4/post/fr2/snapshot_pod_data/x_D_50/fr2_2110100_101/pod_modes/'
file_header = 'low_dim_mode_'

nit = nend_mode - nstart_mode + 1

low_mode = np.zeros((N, nit))

for i in range(nstart_mode , nend_mode + 1):
    filename   =  path + file_header + '%04d'%i + '.txt'
    print(filename)
    low_mode[:,i-nstart_mode] = np.loadtxt(filename)
```

```python
###.......................Reading the Qsub matrix from the binary file.................................. 

nr = 354
ntheta = 256
numvar = 3

Nrows = nr*ntheta*numvar

nstart = 1
nend = 2501

path = '/home/sheel/Work/projects/spod_re5e4/post/frinf/snapshot_pod_data/x_D_100/frinf_2111900_2501/qsub_files/'
#path = '/home/sheel/Work/projects/spod_re5e4/post/fr2/snapshot_pod_data/x_D_50/fr2_2110100_2501/qsub_files/'

file_header = 'Q_sub_'
filename = path + file_header

nit = nend - nstart + 1

Q_sub = np.zeros((Nrows, nit))

for i in range(0, nit):
    num = i + nstart
#     print(num)
    filename   =  path + file_header + '%04d'%num + '.bin'
#    A = np.fromfile(filename, dtype=np.float64)
    f = FortranFile(filename, 'r')          ## Defining the fortran file
    a = f.read_reals(dtype='float64')       ## Reading from fortran file
    Q_sub[:,i] = a
```

```python
print(Q_sub.shape)
print(np.amin(Q_sub))
print(np.amax(Q_sub))
```

```python
###......................Constructing the full field POD modes as Psi = Q_sub * phi.......................

nit = nend_mode - nstart_mode + 1

full_mode = np.zeros((nr*ntheta*numvar, nit))

for i in range(nstart_mode, nend_mode + 1):
    full_mode[:,i-nstart_mode] = np.matmul(Q_sub,low_mode[:,i-nstart_mode])
```

```python
###.........................................Visualize certain POD modes..............................

nr = 354; ntheta = 256

nit = nend_mode - nstart_mode + 1

u_eigenmode = np.zeros((nr, ntheta+2, nit))
v_eigenmode = np.zeros((nr, ntheta+2, nit))
w_eigenmode = np.zeros((nr, ntheta+2, nit))
# rho_eigenmode = np.zeros((nr, ntheta+2, nit))


for i in range(nstart_mode, nend_mode+1):
    u_mode_flat = full_mode[0:nr*ntheta,i-nstart_mode]
    print(np.amin(u_mode_flat)); print(np.amax(u_mode_flat)) 
    u_mode = np.reshape(u_mode_flat,(nr, ntheta), order = 'F')
    
    v_mode_flat = full_mode[nr*ntheta:2*nr*ntheta, i-nstart_mode]
    print(np.amin(v_mode_flat)); print(np.amax(v_mode_flat)) 
    v_mode = np.reshape(v_mode_flat,(nr, ntheta), order = 'F')
    
    w_mode_flat = full_mode[2*nr*ntheta:3*nr*ntheta, i-nstart_mode]
    print(np.amin(w_mode_flat)); print(np.amax(w_mode_flat)) 
    w_mode = np.reshape(w_mode_flat,(nr, ntheta), order = 'F')

#     rho_mode_flat = full_mode[3*nr*ntheta:4*nr*ntheta, i-nstart_mode]
#     print(np.amin(w_mode_flat)); print(np.amax(rho_mode_flat)) 
#     rho_mode = np.reshape(rho_mode_flat,(nr, ntheta), order = 'F')
    
    
    u_eigenmode[:,1:-1,i-nstart_mode] = u_mode  ## Final matrix containing u-velocity eigenmodes
    u_eigenmode[:,0,i-nstart_mode]    = u_mode[:,-1]
    u_eigenmode[:,-1,i-nstart_mode]   = u_mode[:,0]
    
    v_eigenmode[:,1:-1,i-nstart_mode] = v_mode  ## Final matrix containing v-velocity eigenmodes
    v_eigenmode[:,0,i-nstart_mode]    = v_mode[:,-1]
    v_eigenmode[:,-1,i-nstart_mode]   = v_mode[:,0]
    
    w_eigenmode[:,1:-1,i-nstart_mode] = w_mode  ## Final matrix containing w-velocity eigenmodes
    w_eigenmode[:,0,i-nstart_mode]    = w_mode[:,-1]
    w_eigenmode[:,-1,i-nstart_mode]   = w_mode[:,0]

#     rho_eigenmode[:,1:-1,i-nstart_mode] = rho_mode  ## Final matrix containing w-velocity eigenmodes
#     rho_eigenmode[:,0,i-nstart_mode]    = rho_mode[:,-1]
#     rho_eigenmode[:,-1,i-nstart_mode]   = rho_mode[:,0]
```
