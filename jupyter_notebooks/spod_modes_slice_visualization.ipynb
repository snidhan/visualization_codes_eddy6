```python
###.......................Reading the low dimensional mode at different frequencies.....................................

Nblk = 17       # Number of blocks
nfreq_start = 250     
nfreq_end = 290

nit_freq = nfreq_end - nfreq_start + 1     # Number of frequencies

path = '/home/sheel/Work/projects/spod_re5e4/post/fr2/spod_data/x_D_100/nfreq512_novlp384_2110100_2378200/eigenmodes/'

file_header = 'freq_eigenvectors_'

low_mode = np.zeros((Nblk, Nblk, nit_freq), dtype = complex)

for i in range(0, nit_freq):
    
    for k in range(0,Nblk):
        
        freq_num = i + nfreq_start
        mode_num = k + 1
        filename   =  path + file_header + '%04d'%freq_num + '_' +  '%04d'%mode_num + '.txt'
        print(filename)
        file_data = np.loadtxt(filename)
        low_mode[:,k,i] = file_data[:,0] + (1j)*file_data[:,1]
```

```python
###.......................Reading the Q_k matrix from the binary file.................................. 

nr = 519
ntheta = 256
numvar = 3

Nrows = nr*ntheta*numvar

path = '/home/sheel/Work/projects/spod_re5e4/post/fr2/spod_data/x_D_50/nfreq512_novlp384_2110100_2378200/eigenmodes/'
#path = '/home/sheel/Work/projects/spod_re5e4/post/fr2/snapshot_pod_data/x_D_50/fr2_2110100_2501/qsub_files/'

Q_k = np.zeros((Nrows, Nblk, nit_freq), dtype = np.complex)

for i in range(0, nit_freq):
    for j in range(0, Nblk):
        
        freq_num = i + nfreq_start
        mode_num = j + 1
        file_header = 'Q_k_freq_real_'
        filename = path + file_header
        filename   =  path + file_header + '%04d'%freq_num + '_' + '%04d'%mode_num + '.bin'
        print(filename)
        f = FortranFile(filename, 'r')          ## Defining the fortran file
        a = f.read_reals(dtype='float64')       ## Reading from fortran file

        file_header = 'Q_k_freq_imag_'
        filename = path + file_header
        filename   =  path + file_header + '%04d'%freq_num + '_' + '%04d'%mode_num + '.bin'
        print(filename)
        f = FortranFile(filename, 'r')          ## Defining the fortran file
        b = f.read_reals(dtype='float64')       ## Reading from fortran file
        
        Q_k[:,j,i] = a + (1j)*b
```

```python
### ----------------------------------- Constructing the eigenmodes from eigenvectores -----------------------

full_mode = np.zeros((Nrows, Nblk, nit_freq), dtype = np.complex)


for i in range(0, nit_freq):
    print(i)
    full_mode[:,:,i] = np.matmul(Q_k[:,:,i], low_mode[:,:,i])
    
Q_k = None
low_mode = None
```

```python
###.........................................Visualize certain POD modes..............................

nr = 519; ntheta = 256

u_eigenmode = np.zeros((nr, ntheta+2, Nblk, nit_freq), dtype = np.complex)
v_eigenmode = np.zeros((nr, ntheta+2, Nblk, nit_freq), dtype = np.complex)
w_eigenmode = np.zeros((nr, ntheta+2, Nblk, nit_freq), dtype = np.complex)
# rho_eigenmode = np.zeros((nr, ntheta+2, nit))

for i in range(0, nit_freq):
    for j in range(0, Nblk):
        u_mode_flat = full_mode[0:nr*ntheta,j,i]
        print(np.amin(u_mode_flat)); print(np.amax(u_mode_flat)) 
        u_mode = np.reshape(u_mode_flat,(nr, ntheta), order = 'F')
    
        v_mode_flat = full_mode[nr*ntheta:2*nr*ntheta, j, i]
        print(np.amin(v_mode_flat)); print(np.amax(v_mode_flat)) 
        v_mode = np.reshape(v_mode_flat,(nr, ntheta), order = 'F')
    
        w_mode_flat = full_mode[2*nr*ntheta:3*nr*ntheta, j, i]
        print(np.amin(w_mode_flat)); print(np.amax(w_mode_flat)) 
        w_mode = np.reshape(w_mode_flat,(nr, ntheta), order = 'F')

#     rho_mode_flat = full_mode[3*nr*ntheta:4*nr*ntheta, i-nstart_mode]
#     print(np.amin(w_mode_flat)); print(np.amax(rho_mode_flat)) 
#     rho_mode = np.reshape(rho_mode_flat,(nr, ntheta), order = 'F')
    
    
        u_eigenmode[:,1:-1,j,i] = u_mode  ## Final matrix containing u-velocity eigenmodes
        u_eigenmode[:,0,j,i]    = u_mode[:,-1]
        u_eigenmode[:,-1,j,i]   = u_mode[:,0]
    
        v_eigenmode[:,1:-1,j,i] = v_mode  ## Final matrix containing v-velocity eigenmodes
        v_eigenmode[:,0,j,i]    = v_mode[:,-1]
        v_eigenmode[:,-1,j,i]   = v_mode[:,0]
    
        w_eigenmode[:,1:-1,j,i] = w_mode  ## Final matrix containing w-velocity eigenmodes
        w_eigenmode[:,0,j,i]    = w_mode[:,-1]
        w_eigenmode[:,-1,j,i]   = w_mode[:,0]

#     rho_eigenmode[:,1:-1,i-nstart_mode] = rho_mode  ## Final matrix containing w-velocity eigenmodes
#     rho_eigenmode[:,0,i-nstart_mode]    = rho_mode[:,-1]
#     rho_eigenmode[:,-1,i-nstart_mode]   = rho_mode[:,0]

```
